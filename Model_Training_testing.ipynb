# LIST OF ALL ATTRIBUTES


```python
onus_attributes = [col for col in dev_dataset.columns if col.startswith("onus")]
transaction_attributes = [col for col in dev_dataset.columns if col.startswith("transaction")]
bureau_attributes = [col for col in dev_dataset.columns if col.startswith("bureau")] ## it is containing all bureau
bureau_enquiry_attributes = [col for col in dev_dataset.columns if col.startswith("bureau_enquiry")]
```

# DATA PREPROCESSING

```python
dataset=dev_dataset.iloc[ : ,1:1215]
```

```python
dataset.head(5)
```

```python
dataset.isnull().sum()
```

# FILLING NULL VALUES 

```python
# Fill missing values in transaction attributes with the mean
for column in transaction_attributes:
    dataset[column].fillna(dataset[column].mean(), inplace=True)

# Fill missing values in the first 452 bureau attributes with the mean
for column in bureau_attributes[0:452]:
    dataset[column].fillna(dataset[column].mean(), inplace=True)

# Fill missing values in bureau enquiry attributes with the mean
for column in bureau_enquiry_attributes:
    dataset[column].fillna(dataset[column].mean(), inplace=True)

# Fill missing values in onus attributes with the mean
for column in onus_attributes:
    dataset[column].fillna(dataset[column].mean(), inplace=True)
```

```python
dataset.isnull().sum()
```

# REMOVE FEATURE USING STANDAR SCALER

```python
from sklearn.feature_selection import VarianceThreshold
import joblib

# Fit the selector on the training data
selector = VarianceThreshold(threshold=0.005)
selector.fit(dataset)

# Save the fitted selector
joblib.dump(selector, 'variance_threshold_selector.pkl')

# Transform the training data
X_train_reduced = selector.transform(dataset)


# Get the names of the remaining features
reduced_columns = dataset.columns[selector.get_support()]
print(f"Remaining features after variance threshold: {len(reduced_columns)}")
print(reduced_columns)
```

```python
reduced_data = dataset[reduced_columns]
```

# OUTLIER REMOVAL USING ISOLATION FOREST

```python
from sklearn.ensemble import IsolationForest
import pandas as pd

# Load your data (example: tran_dataset)
# Assuming tran_dataset is a DataFrame with only features (no target variable yet)
X = reduced_data  # Drop target column if present

# Initialize the IsolationForest model
model = IsolationForest(n_estimators=200, contamination='auto', random_state=15)

# Fit the model on your dataset
model.fit(X)

# Predict outliers (anomalies)
# The output is an array where -1 indicates an outlier and 1 indicates a normal data point
outlier_predictions = model.predict(X)

# Create a new column to store the outlier predictions using .loc to avoid the SettingWithCopyWarning
reduced_data.loc[:, 'outlier'] = outlier_predictions

# If you want to only keep non-outliers:
dataset_no_outliers = reduced_data[reduced_data['outlier'] == 1].copy()

# Check the results (how many outliers were detected)
print(f"Number of outliers detected: {sum(outlier_predictions == -1)}")
print(f"Number of non-outliers: {sum(outlier_predictions == 1)}")

# You can now proceed with further analysis on the 'dataset_no_outliers'
```

# TRAIN TEST AND BALANCE THE DATASET

```python
input=transform_dataset.iloc[:,0:937]
output=transform_dataset.iloc[:,937:]

```

# BALANCING CLASS  

```python
from imblearn.over_sampling import RandomOverSampler
from sklearn.model_selection import train_test_split

# Assuming 'input' is your feature matrix and 'output' is your target variable
# Apply Random Oversampling (ROS) to the training data
ros = RandomOverSampler(random_state=15)

# Resample the data using Random Oversampling
X_train_ros, y_train_ros = ros.fit_resample(input, output)

# Check class distribution after ROS
print("Original class distribution:\n", output.value_counts())
print("After ROS class distribution:\n", y_train_ros.value_counts())

# Now, you can use X_train_ros and y_train_ros to train your model
```

# SPLITTING DATA

```python
from sklearn.model_selection import train_test_split

# Split the resampled data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(
    X_train_ros, y_train_ros, test_size=0.3, random_state=15, stratify=y_train_ros
)

# Check the shape of your training and testing data
print(f"Training data shape: {X_train.shape}")
print(f"Testing data shape: {X_test.shape}")
```
